```python
plt.figure(figsize=(8,5))
df['Crop'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Crop Distribution')
plt.xlabel('Crop')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Season'].value_counts().plot(kind='bar', color='blue')
plt.title('Season Distribution')
plt.xlabel('Season')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['State'].value_counts().plot(kind='bar', color='darkblue')
plt.title('State Distribution')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Yield'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Yield Distribution')
plt.xlabel('Yield')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Area'], df['Production'], color='green')
plt.title('Production vs Area')
plt.xlabel('Area')
plt.ylabel('Production')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Annual_Rainfall'], df['Production'], color='purple')
plt.title('Production vs Annual Rainfall')
plt.xlabel('Annual Rainfall)')
plt.ylabel('Production')
plt.show()
```

```python
plt.figure(figsize=(30,20))
plt.bar(df['State'], df['Production'], color='yellow', edgecolor='black')
plt.title('Production vs State')
plt.xlabel('State')
plt.ylabel('Production')
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Fertilizer'], df['Yield'], label='Fertilizer', color='blue')
plt.scatter(df['Pesticide'], df['Yield'], label='Pesticide', color='red')
plt.title('Yield vs Fertilizer and Pesticide')
plt.xlabel('Fertilizer / Pesticide (kg per hectare)')
plt.ylabel('Yield (tons per hectare)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Yield'], df['Production'], color='purple', edgecolor='black')
plt.title('Production vs Yield')
plt.xlabel('Yield')
plt.ylabel('Production')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Area'], df['Yield'], color='green')
plt.title('Yield vs Area')
plt.xlabel('Area')
plt.ylabel('Yield')
plt.show()
```

```python
plt.figure(figsize=(30,20))
plt.bar(df['State'], df['Yield'], color='yellow')
plt.title('Yield vs State')
plt.xlabel('State')
plt.ylabel('Yield')
```
