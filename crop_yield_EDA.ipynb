```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('crop_yield.csv')
```

```python
plt.figure(figsize = (10,9))
sns.countplot(y="Crop",data=df, order=df['Crop'].value_counts().index)
plt.title("Crop Distribution")
plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.countplot(y="Season",data=df, order=df['Season'].value_counts().index)
plt.title("Season Distribution")
plt.show()
```

```python
plt.figure(figsize = (10,9))
sns.countplot(y="State",data=df, order=df['State'].value_counts().index)
plt.title("State Distribution")
plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.histplot(df['Yield'], kde=True, bins=30)
plt.title("Yield Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x='Area', y='Production', hue='Crop', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Production vs Area")
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x='Fertilizer', y='Yield', hue='Crop', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Yield vs Fertilizer ")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='Annual_Rainfall', y='Production', hue='State', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Production vs Annual Rain by State")
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.lineplot(x='Crop_Year', y='Yield', hue='Crop', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Yield Trends Over the Years by Crop")
plt.show()
```
